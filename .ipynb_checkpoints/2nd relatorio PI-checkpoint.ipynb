# Add novas colunas de preco, custo Unitario e impacto de Precos e Volume

```python
# Add Unit Cost for 2023
df['Unit Cost 2023'] = df.apply(
    lambda row: row['Total Cost Actual 2023'] / row['Qtd Actual 2023'] 
    if row['Qtd Actual 2023'] != 0 else 0, 
    axis=1
)

# Add Unit Cost for 2022
df['Unit Cost 2022'] = df.apply(
    lambda row: row['Total Cost Actual 2022'] / row['Qtd Actual 2022'] 
    if row['Qtd Actual 2022'] != 0 else 0, 
    axis=1
)

# Add Unit Cost for 2024
df['Unit Cost 2024'] = df.apply(
    lambda row: row['Total Cost Plan 2024'] / row['Qtd Plan 2024'] 
    if row['Qtd Plan 2024'] != 0 else 0, 
    axis=1
)

# Add Unit Cost for 2025
df['Unit Cost 2025'] = df.apply(
    lambda row: row['Total Cost Plan 2025'] / row['Qtd Plan 2025'] 
    if row['Qtd Plan 2025'] != 0 else 0, 
    axis=1
)
```

# add novas colunas de Impactos financeiros

```python
# Add a new column with the condition
df['Mix Status 23 vs 22'] = df.apply(
    lambda row: 'MIX' if row['Qtd Actual 2023'] == 0 or row['Qtd Actual 2022'] == 0 else 'NOT MIX', 
    axis=1
)

# Add a new column with the condition
df['Mix Status 24 vs 23'] = df.apply(
    lambda row: 'MIX' if row['Qtd Plan 2024'] == 0 or row['Qtd Actual 2023'] == 0 else 'NOT MIX', 
    axis=1
)

# Add a new column with the condition
df['Mix Status 25 vs 24'] = df.apply(
    lambda row: 'MIX' if row['Qtd Plan 2024'] == 0 or row['Qtd Plan 2025'] == 0 else 'NOT MIX', 
    axis=1
)
 # Add a new column with the condition for Price Impact, excluding MIX rows
df['Price Impact 23 vs 22'] = df.apply(
    lambda row: row['Qtd Actual 2023'] * (row['Price Actual 2023'] - row['Price Actual 2022'])
    if row['Mix Status 23 vs 22'] != 'MIX' else 0, 
    axis=1
)
 # Add a new column with the condition for Price Impact, excluding MIX rows
df['Price Impact 24 vs 23'] = df.apply(
    lambda row: row['Qtd Plan 2024'] * (row['Price Plan 2024'] - row['Price Actual 2023'])
    if row['Mix Status 24 vs 23'] != 'MIX' else 0, 
    axis=1
)
 # Add a new column with the condition for Price Impact, excluding MIX rows
df['Price Impact 25 vs 24'] = df.apply(
    lambda row: row['Qtd Plan 2025'] * (row['Price Plan 2025'] - row['Price Plan 2024'])
    if row['Mix Status 25 vs 24'] != 'MIX' else 0, 
    axis=1
)

# Add a new column with the condition for Volume Impact, excluding MIX rows
df['Volume Impact 23 vs 22'] = df.apply(
    lambda row: row['Price Actual 2022']*(row['Qtd Actual 2023'] - row['Qtd Actual 2022'] )
    if row['Mix Status 23 vs 22'] != 'MIX' else 0, 
    axis=1
)

# Add a new column with the condition for Volume Impact, excluding MIX rows
df['Volume Impact 24 vs 23'] = df.apply(
    lambda row: row['Price Actual 2023']*(row['Qtd Plan 2024'] - row['Qtd Actual 2023'] )
    if row['Mix Status 24 vs 23'] != 'MIX' else 0, 
    axis=1
)

 # Add a new column with the condition for Volume Impact, excluding MIX rows
df['Volume Impact 25 vs 24'] = df.apply(
    lambda row: row['Price Plan 2024']*(row['Qtd Plan 2025'] - row['Qtd Plan 2024'] )
    if row['Mix Status 25 vs 24'] != 'MIX' else 0, 
    axis=1
)
```

```python
import matplotlib.pyplot as plt
sales_by_region = df.groupby('Region')[['Sales Actual 2022', 'Sales Actual 2023', 'Sales Plan 2024']].sum()
# Plotting
sales_by_region.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show() 
```
